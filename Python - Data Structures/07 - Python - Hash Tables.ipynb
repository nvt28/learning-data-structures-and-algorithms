## Hash Tables

- A hash table (or hash map) is a data structure that implements an associative array abstract data type, a structure that can map keys to values.
- A hash table uses a hash function to compute an index, also called a hash code, into an array of buckets or slots, from which the desired value can be found

In general, hash tables store key-value pairs and the key is generated using a hash function.

#### Functions
The functions associated with Hash table are:

|Function | Description | Average Case | Worst case |
|---|---|---|---|
|`seach(X)`|Checks whether the number is present in the table or not| O(1) | O(n) |
|`insert(X)`|Adds the element 'X' in the Hash table at mapped to the key calculated by the hash function | O(1) | O(n) |
|`delete(X)`|Deletes the element 'X' from the Hash table | O(1) | O(n) |

Worst case complexities are only possible when the Hash function is designed badly and maps all elements to the same key

### Using `Dict` as a Hash table

- Hash tables or hash maps in Python are implemented through the built-in dictionary data type
- The keys of a dictionary in Python are generated by a hashing function
- The elements of a dictionary are not ordered and they can be changed

Note: As of Python version 3.7, dictionaries are **ordered**. In Python 3.6 and earlier, dictionaries are unordered.

- Dictionaries can be created in two ways:
    - Using curly braces ({})
    - Using the `dict()` function

```python
new_dict = dict(Dave = '001' , Ava= '002' , Joe= '003')
new_dict_2 = {}

print(new_dict)
print(type(new_dict), type(new_dict_2))
```

#### Dictionary Operations
- Elements can be added to the dict using a new key and assigning it a value: `dict[key] = value`
- Elements can be accessed using `dict.get(key)` or `dict[key]`
- Elements can be deleted using `dict.pop(key)` or the `del` keyword with specified key `del dict[key]`
- To check if a specified key is present in a dict, use the `in` keyword
- Empty stack can be found using `len(dict)`
